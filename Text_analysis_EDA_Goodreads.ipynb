#Text-Preprocessing

```python
#Pre-processing for Train set
#stopwords, punctuation, url removal, lower casing,
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('stopwords') #downloading stopwords

stop_words = set(stopwords.words('english')) #english stopwords

GRTrain['cleaned_review'] = GRTrain['review_text'].str.lower().str.replace('[^\w\s]','').apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))

def remove_urls(cleaned_review):
    return re.sub(r'http\S+', '', cleaned_review)
GRTrain['cleaned_review'] = GRTrain['cleaned_review'].apply(remove_urls)
```

```python
#Tokenization
def tokenization (cleaned_review):
  tokens=re.split(r'\W+', cleaned_review)
  return tokens
GRTrain['cleaned_review'] = GRTrain['cleaned_review'].apply(tokenization)
```

```python
import nltk
nltk.download('wordnet')
```

```python
#stemming and lemmatization
def stemming(text):
    stemmer = PorterStemmer()
    return [stemmer.stem(word) for word in text]
GRTrain['cleaned_review'] = GRTrain['cleaned_review'].apply(stemming)
```

```python
def lemmatize(text):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word) for word in text]
GRTrain['cleaned_review'] = GRTrain['cleaned_review'].apply(lemmatize)
```
